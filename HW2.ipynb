1.sınıf, 2.sınıf ve 3.sınıf yolcuların ölüm oranlarını karşılaştırınız. Bir veya bir kaç cümle ile sonucu <b>yorumlayınız.</b>


```python
# bu alana python kodunu yazıp çıktısı görülecek şekilde çalıştırınız
print("Yorumunuzu print() fonksiyonu içine yazabilirsiniz.")
```

```python
display(titanic_df)
```

```python
p1 = titanic_df[(titanic_df.Pclass == 1)]
p2 = titanic_df[(titanic_df.Pclass == 2)]
p3 = titanic_df[(titanic_df.Pclass == 3)]

d1 = titanic_df[(titanic_df.Pclass == 1) & (titanic_df.Survived == 0)]
d2 = titanic_df[(titanic_df.Pclass == 2) & (titanic_df.Survived == 0)]
d3 = titanic_df[(titanic_df.Pclass == 3) & (titanic_df.Survived == 0)]

s1 = len(d1) / len(p1)
s2 = len(d2) / len(p2)
s3 = len(d3) / len(p3)

print("1.Sınıf Yolcu Ölüm Oranı: ", s1)
print("2.Sınıf Yolcu Ölüm Oranı: ", s2)
print("3.Sınıf Yolcu Ölüm Oranı: ", s3)
print("\niyi sınıfların (1'den 3'e doğru) ölüm oranı daha düşük.\n")
```

# Missing Value Doldurma

Bundan sonraki soruları <b>train_missing.csv</b> datasını kullanarak cevaplayınız.

# Soru 8

```python
# bu alana python kodunu yazıp çıktısı görülecek şekilde çalıştırınız
missing_df = pd.read_csv("train_missing.csv")
display(missing_df)
```

Age, Embarked ve Fare değişkenlerini mean/median/mode değerlerini kullanarak değişken tipine uygun olacak şekilde doldurunuz.

```python
missing_df.isnull().sum()
```

```python
missing_df["Age"].median()
```

```python
missing_df["Age"]  = missing_df["Age"].fillna(missing_df["Age"].median())
display(missing_df["Age"])
```

```python
missing_df["Embarked"].mode()
```

```python
missing_df["Embarked"] = missing_df["Embarked"].fillna(missing_df["Embarked"].mode()[0])
display(missing_df)
```

```python
missing_df["Fare"].mean()
```

```python
missing_df["Fare"] = missing_df["Fare"].fillna(missing_df["Fare"].mean())
display(missing_df)
```
